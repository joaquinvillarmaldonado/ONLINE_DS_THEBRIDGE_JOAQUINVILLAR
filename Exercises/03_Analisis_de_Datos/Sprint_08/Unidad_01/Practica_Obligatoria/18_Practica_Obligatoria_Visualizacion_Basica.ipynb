### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# 1. Ciudad más pobladas (con texto en el gráfico)
plt.figure(figsize=(8, 6))

# Seleccionar las ciudades con mayor población
top_cities = df_cleaned[['city', 'population_total']].sort_values(by='population_total', ascending=False).head(10)

# Usar 10 colores para las 10 ciudades
colors = ['blue', 'gray', 'olive', 'brown', 'blue', 'green', 'red', 'purple', 'orange', 'pink']

# Crear el gráfico de dispersión con los colores correctos
plt.figure(figsize=(8, 6))
plt.scatter(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.5)
for i, row in top_cities.iterrows():
    plt.text(row['city'], row['population_total'], str(row['population_total']), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 ciudades más pobladas")
plt.xlabel("Ciudad")
plt.ylabel("Población Total")
plt.tight_layout()
plt.show()
```

```python
# 2. Histograma del área total ocupada por las ciudades
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned['area_total_sq_mi'], bins=50, color='steelblue')
plt.title("Distribución del área total de las ciudades")
plt.xlabel("Área Total (millas²)")
plt.ylabel("Frecuencia")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter plot de área de agua vs población
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['area_water_sq_mi'], df_cleaned['population_total'], alpha=0.5, color='green')
plt.title("Área de agua vs Población de las ciudades")
plt.xlabel("Área de agua (millas²)")
plt.ylabel("Población Total")
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplot de la altura de las ciudades
plt.figure(figsize=(8, 6))
plt.boxplot(df_cleaned['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightcoral"))
plt.title("Boxplot de la altura de las ciudades")
plt.xlabel("Altura (metros)")
plt.tight_layout()
plt.show()
```

```python
# Crear una figura con 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Las ciudades más pobladas
top_cities = df_cleaned[['city', 'population_total']].sort_values(by='population_total', ascending=False).head(10)
colors = ['blue', 'gray', 'olive', 'brown', 'blue', 'green', 'red', 'purple', 'orange', 'pink']

axs[0, 0].scatter(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.5)
for i, row in top_cities.iterrows():
    axs[0, 0].text(row['city'], row['population_total'], str(row['population_total']), ha='center', va='bottom', fontsize=10)
axs[0, 0].set_title("Top 10 ciudades más pobladas")
axs[0, 0].set_xlabel("Ciudad")
axs[0, 0].set_ylabel("Población Total")
axs[0, 0].tick_params(axis='x', rotation=45)

# 2. Histograma del área total ocupada por las ciudades
axs[0, 1].hist(df_cleaned['area_total_sq_mi'], bins=50, color='steelblue')
axs[0, 1].set_title("Histograma del área total ocupada")
axs[0, 1].set_xlabel("Área Total (millas cuadradas)")
axs[0, 1].set_ylabel("Frecuencia")
axs[0, 1].tick_params(axis='x', rotation=45)

# 3. Scatter de área de agua vs población
axs[1, 0].scatter(df_cleaned['area_water_sq_mi'], df_cleaned['population_total'], color='green', alpha=0.5)
axs[1, 0].set_title("Área de agua vs Población")
axs[1, 0].set_xlabel("Área de agua (millas cuadradas)")
axs[1, 0].set_ylabel("Población Total")

# 4. Boxplot de la altura de las ciudades
axs[1, 1].boxplot(df_cleaned['elevation_m'].dropna(), vert=False, patch_artist=True)
axs[1, 1].set_title("Boxplot de la altura de las ciudades")
axs[1, 1].set_xlabel("Altura (metros)")

# Ajustar el espaciado entre subgráficos
plt.tight_layout()

# Mostrar la figura completa con los cuatro gráficos
plt.show()
```
