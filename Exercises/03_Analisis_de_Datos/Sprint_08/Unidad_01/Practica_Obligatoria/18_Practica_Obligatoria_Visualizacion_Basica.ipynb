### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# 1. Ciudad ms pobladas (con texto en el grfico)
plt.figure(figsize=(8, 6))

# Seleccionar las ciudades con mayor poblacin
top_cities = df_cleaned[['city', 'population_total']].sort_values(by='population_total', ascending=False).head(10)

# Usar 10 colores para las 10 ciudades
colors = ['blue', 'gray', 'olive', 'brown', 'blue', 'green', 'red', 'purple', 'orange', 'pink']

# Crear el grfico de dispersin con los colores correctos
plt.figure(figsize=(8, 6))
plt.scatter(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.5)
for i, row in top_cities.iterrows():
    plt.text(row['city'], row['population_total'], str(row['population_total']), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 ciudades ms pobladas")
plt.xlabel("Ciudad")
plt.ylabel("Poblacin Total")
plt.tight_layout()
plt.show()
```

```python
# 2. Histograma del rea total ocupada por las ciudades
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned['area_total_sq_mi'], bins=50, color='steelblue')
plt.title("Distribucin del rea total de las ciudades")
plt.xlabel("rea Total (millas)")
plt.ylabel("Frecuencia")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. Scatter plot de rea de agua vs poblacin
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['area_water_sq_mi'], df_cleaned['population_total'], alpha=0.5, color='green')
plt.title("rea de agua vs Poblacin de las ciudades")
plt.xlabel("rea de agua (millas)")
plt.ylabel("Poblacin Total")
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplot de la altura de las ciudades
plt.figure(figsize=(8, 6))
plt.boxplot(df_cleaned['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightcoral"))
plt.title("Boxplot de la altura de las ciudades")
plt.xlabel("Altura (metros)")
plt.tight_layout()
plt.show()
```

```python
# Crear una figura con 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Las ciudades ms pobladas
top_cities = df_cleaned[['city', 'population_total']].sort_values(by='population_total', ascending=False).head(10)
colors = ['blue', 'gray', 'olive', 'brown', 'blue', 'green', 'red', 'purple', 'orange', 'pink']

axs[0, 0].scatter(top_cities['city'], top_cities['population_total'], color=colors, alpha=0.5)
for i, row in top_cities.iterrows():
    axs[0, 0].text(row['city'], row['population_total'], str(row['population_total']), ha='center', va='bottom', fontsize=10)
axs[0, 0].set_title("Top 10 ciudades ms pobladas")
axs[0, 0].set_xlabel("Ciudad")
axs[0, 0].set_ylabel("Poblacin Total")
axs[0, 0].tick_params(axis='x', rotation=45)

# 2. Histograma del rea total ocupada por las ciudades
axs[0, 1].hist(df_cleaned['area_total_sq_mi'], bins=50, color='steelblue')
axs[0, 1].set_title("Histograma del rea total ocupada")
axs[0, 1].set_xlabel("rea Total (millas cuadradas)")
axs[0, 1].set_ylabel("Frecuencia")
axs[0, 1].tick_params(axis='x', rotation=45)

# 3. Scatter de rea de agua vs poblacin
axs[1, 0].scatter(df_cleaned['area_water_sq_mi'], df_cleaned['population_total'], color='green', alpha=0.5)
axs[1, 0].set_title("rea de agua vs Poblacin")
axs[1, 0].set_xlabel("rea de agua (millas cuadradas)")
axs[1, 0].set_ylabel("Poblacin Total")

# 4. Boxplot de la altura de las ciudades
axs[1, 1].boxplot(df_cleaned['elevation_m'].dropna(), vert=False, patch_artist=True)
axs[1, 1].set_title("Boxplot de la altura de las ciudades")
axs[1, 1].set_xlabel("Altura (metros)")

# Ajustar el espaciado entre subgrficos
plt.tight_layout()

# Mostrar la figura completa con los cuatro grficos
plt.show()
```
